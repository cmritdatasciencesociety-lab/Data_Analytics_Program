# Line Plot

## Purpose

A Line Plot is used to visualize trends and changes in data over time or across an ordered sequence of observations.

---

## Real-World Scenario

Imagine you own an ice cream shop.

Every month, you track the number of ice creams sold. After a year, you want to understand:

* During which months do customers buy the most ice cream?
* How do seasons affect sales?
* Which months performed best?
* Are there any unexpected drops in sales?

Instead of looking at a table of numbers, a Line Plot allows you to visualize the sales trend clearly.

---

## When Should We Use It?

Use a Line Plot when:

* Data is collected over time (days, weeks, months, years)
* You want to identify trends and patterns
* You need to monitor growth or decline
* The order of observations is important

---

## Common Applications

* Ice Cream Sales by Month
* Monthly Sales Revenue
* Website Traffic Growth
* App Downloads
* Stock Prices
* Temperature Changes
* Social Media Followers Over Time
* Daily Active Users (DAU)

---

## Question Answered

**"How is a value changing over time?"**

---

## Example

Suppose an ice cream shop records the following monthly sales:

| Month | Ice Creams Sold |
| ----- | --------------- |
| Jan   | 80              |
| Feb   | 100             |
| Mar   | 150             |
| Apr   | 220             |
| May   | 300             |
| Jun   | 380             |
| Jul   | 420             |
| Aug   | 400             |
| Sep   | 320             |
| Oct   | 220             |
| Nov   | 140             |
| Dec   | 90              |

A Line Plot helps us quickly observe:

* Sales increase as the weather gets warmer
* Peak sales occur during summer months
* Sales decline as winter approaches
* Seasonal patterns in customer demand

We will now create a Line Plot using Matplotlib to visualize these monthly ice cream sales.


````python
import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun",
          "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"]

ice_cream_sales = [80, 100, 150, 220, 300, 380,
                   420, 400, 320, 220, 140, 90]

plt.plot(months, ice_cream_sales)

plt.title("Monthly Ice Cream Sales Trend")
plt.xlabel("Month")
plt.ylabel("Ice Cream Sales")
plt.grid()

plt.show()
````

Is sales increasing or decreasing?

Which month has highest sales?

Is growth consistent?

# Histogram

## Purpose

A Histogram is used to understand the distribution of numerical data.

Unlike a Bar Plot, a Histogram does not compare categories. Instead, it groups numerical values into ranges called **bins** and shows how many observations fall into each range.

---

## Real-World Scenario

Imagine the HR department of a company wants to analyze employee salaries.

The company has employees with different levels of experience, and management wants to understand:

* What salary range contains the majority of employees?
* Are most employees earning similar salaries?
* Are there a few employees earning exceptionally high salaries?
* How are salaries distributed across the organization?

Looking at hundreds of salary records individually can be difficult. A Histogram helps summarize the distribution by grouping salaries into ranges.

---

## When Should We Use It?

Use a Histogram when you want to know:

* Where values are concentrated
* How data is spread
* Whether unusual values (outliers) exist
* The overall shape of the data distribution

---

## How Does a Histogram Work?

A Histogram automatically divides numerical data into intervals called **bins**.

For example, if employee salaries range from ₹20,000 to ₹1,00,000, the Histogram may create bins such as:

```text
20,000 - 40,000
40,000 - 60,000
60,000 - 80,000
80,000 - 1,00,000
```

It then counts how many employees fall into each salary range and displays the frequencies as bars.

---

## Can We Create Our Own Bins?

Yes.

By default, Matplotlib automatically decides the number of bins.

However, we can control the bins ourselves.

For example:

```python
plt.hist(salaries, bins=5)
```

This creates 5 bins.

We can also define custom bin ranges:

```python
plt.hist(salaries, bins=[20000,40000,60000,80000,100000])
```

This creates the following bins:

```text
20,000 - 40,000
40,000 - 60,000
60,000 - 80,000
80,000 - 1,00,000
```

This gives us more control over how the data is grouped and analyzed.

---

## Connection to Statistics

Previously we learned:

* Mean
* Median
* Range
* Quartiles
* IQR

A Histogram helps us visualize these concepts and understand the overall distribution of data.

---

## Common Applications

* Employee Salary Distribution
* Student Marks Distribution
* Age Distribution
* Customer Spending Patterns
* House Prices
* Employee Experience Distribution

---

## Question Answered

**"How is the data distributed?"**

---

## Example

Suppose a company records the monthly salaries of its employees.

A Histogram can help us identify:

* Which salary range contains the most employees
* Whether salaries are concentrated in a specific range
* Whether there are unusually high salaries
* How evenly salaries are distributed

Instead of looking at individual salary values, the Histogram groups them into ranges and makes patterns easier to understand.

---

## What Can We Learn From a Histogram?

A Histogram helps us identify:

* Where most values are concentrated
* Whether the data is symmetric or skewed
* The spread of the data
* Possible outliers
* Gaps in the data
* The overall shape of the distribution

We will now create a Histogram using Matplotlib to visualize the distribution of employee salaries.


````python
import matplotlib.pyplot as plt

salaries = [25000, 28000, 30000, 32000, 35000,
            38000, 40000, 42000, 45000, 48000,
            50000, 52000, 55000, 58000, 60000,
            65000, 70000, 75000, 85000, 100000]

plt.hist(salaries)

plt.title("Employee Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")

plt.show()
````

# Scatter Plot

## Purpose

A Scatter Plot is used to understand the relationship between two numerical variables.

It helps us determine whether changes in one variable are associated with changes in another variable.

---

## Real-World Scenario

Imagine you run an online store and spend money on digital advertising every month.

At the end of the year, you want to understand:

* Does spending more on advertisements increase sales?
* Is there a relationship between advertising budget and revenue?
* Are there any unusual months where high spending did not generate high sales?

Instead of looking at tables of numbers, a Scatter Plot helps visualize the relationship between these two variables.

---

## When Should We Use It?

Use a Scatter Plot when:

* You have two numerical variables
* You want to understand whether they are related
* You want to identify patterns or trends
* You want to detect unusual observations (outliers)

---

## Common Applications

* Advertising Spend vs Sales
* Experience vs Salary
* Study Hours vs Exam Marks
* Temperature vs Ice Cream Sales
* House Size vs House Price
* Website Visitors vs Revenue

---

## Question Answered

**"Are two variables related?"**

---

## Example

Suppose an online store records the following data:

| Advertising Spend (₹) | Sales (₹) |
| --------------------- | --------- |
| 10,000                | 80,000    |
| 15,000                | 1,00,000  |
| 20,000                | 1,30,000  |
| 25,000                | 1,55,000  |
| 30,000                | 1,80,000  |

Each point on the Scatter Plot represents one month's advertising spend and corresponding sales.

By observing the plot, we can determine whether increased advertising spending is associated with increased sales.

---

## Relationship Types

### Positive Correlation

As one variable increases, the other variable also increases.

**Example:** Advertising Spend vs Sales

```text
More Ads → More Sales
```

---

### Negative Correlation

As one variable increases, the other variable decreases.

**Example:** Product Price vs Number of Customers

```text
Higher Price → Fewer Customers
```

---

### No Correlation

No clear relationship exists between the variables.

**Example:** Employee ID vs Monthly Sales

```text
Changes in one variable do not affect the other.
```

---

## What Can We Learn From a Scatter Plot?

A Scatter Plot helps us identify:

* Whether two variables are related
* The direction of the relationship (positive or negative)
* The strength of the relationship
* Unusual observations (outliers)
* Patterns that may not be visible in a table of data

We will now create a Scatter Plot using Matplotlib to visualize the relationship between advertising spend and sales.


````python


ad_spend = [10000, 15000, 20000, 25000, 30000, 35000, 40000, 45000]

sales = [80000, 100000, 130000, 155000, 180000, 210000, 240000, 270000]

plt.scatter(ad_spend, sales)

plt.title("Advertising Spend vs Sales")
plt.xlabel("Advertising Spend (₹)")
plt.ylabel("Sales Revenue (₹)")

plt.show()
````

# Pie Chart

## Purpose

A Pie Chart shows how different categories contribute to a whole.

Each slice represents a proportion or percentage of the total.

---

## Real-World Scenario

Imagine you are managing your monthly budget.

At the end of the month, you want to understand:

* How much money was spent on rent?
* How much was spent on food?
* How much went towards transportation?
* Which category consumed the largest portion of your budget?

Instead of looking at a table of expenses, a Pie Chart helps visualize how each expense category contributes to the total monthly spending.

---

## When Should We Use It?

Use a Pie Chart when:

* You want to show parts of a whole
* Percentages or proportions are important
* You want to compare contribution rather than exact values

---

## Common Applications

* Monthly Expense Distribution
* Market Share Analysis
* Budget Allocation
* Revenue Contribution by Product
* Customer Segmentation
* Department-wise Budget Usage

---

## Question Answered

**"How much does each category contribute?"**

---

## Example

Suppose a person's monthly expenses are distributed as follows:

| Category       | Amount (₹) |
| -------------- | ---------- |
| Rent           | 20,000     |
| Food           | 8,000      |
| Transportation | 4,000      |
| Entertainment  | 3,000      |
| Savings        | 15,000     |

A Pie Chart helps us quickly observe:

* Which category consumes the largest portion of the budget
* Which category contributes the least
* The percentage contribution of each category
* How the total budget is distributed

---

## Important Note

Pie Charts work best when:

* The number of categories is small
* The categories collectively represent a whole
* The total equals 100% (or represents a complete total)

Too many slices make interpretation difficult and can reduce readability.

---

## When Not to Use a Pie Chart

Avoid Pie Charts when:

* There are too many categories
* Precise comparisons are important
* Categories do not represent parts of a whole

In such cases, a Bar Plot is often a better choice.

We will now create a Pie Chart using Matplotlib to visualize the distribution of monthly expenses.


````python
categories = ["Rent", "Food", "Transportation", "Entertainment", "Savings"]

amounts = [20000, 8000, 4000, 3000, 15000]

plt.pie(amounts, labels=categories, autopct="%1.1f%%")

plt.title("Monthly Expense Distribution")

plt.show()
````

# Box Plot

## Purpose

A Box Plot summarizes the distribution of numerical data using key statistical measures.

It helps us understand the spread of the data and identify unusual values (outliers).

---

## Real-World Scenario

Imagine the HR department of a company wants to analyze employee salaries.

The company has employees ranging from fresh graduates to senior managers. Management wants to understand:

* What is the typical salary range?
* How spread out are employee salaries?
* Are there employees earning unusually high or low salaries?
* How does the salary distribution look overall?

Instead of examining hundreds of salary records individually, a Box Plot provides a quick summary of the distribution.

---

## When Should We Use It?

Use a Box Plot when you want to:

* Detect outliers
* Compare distributions
* Understand the spread of data
* Visualize key statistical measures

---

## Connection to Statistics

A Box Plot is built using:

* Minimum
* First Quartile (Q1)
* Median (Q2)
* Third Quartile (Q3)
* Maximum

It also highlights potential outliers using the IQR (Interquartile Range) method.

---

## Example

Suppose a company records the monthly salaries (₹) of its employees.

A Box Plot can help us identify:

* The middle 50% of salaries
* The median salary
* The overall spread of salaries
* Employees whose salaries are unusually high or low

For example, if most employees earn between ₹30,000 and ₹70,000 but a few executives earn ₹2,50,000+, those salaries may appear as outliers on the Box Plot.

---

## Question Answered

**"Are there unusual values in the data?"**

---

## What Can We Learn From a Box Plot?

A Box Plot helps us identify:

* The center of the data (Median)
* The spread of the data (IQR)
* The range of typical values
* Potential outliers
* Whether the distribution is symmetric or skewed

---

## Why Is It Important?

Box Plots are widely used in Data Analytics because they provide a quick statistical summary without displaying every individual data point.

They are especially useful when comparing multiple groups or detecting unusual observations that may require further investigation.

We will now create a Box Plot using Matplotlib to visualize employee salary distribution.


````python
import matplotlib.pyplot as plt

salaries = [25000, 28000, 30000, 32000, 35000,
            38000, 40000, 42000, 45000, 48000,
            50000, 52000, 55000, 58000, 60000,
            65000, 70000, 75000, 85000, 250000,200000
           ]

plt.boxplot(salaries)

plt.title("Employee Salary Distribution")
plt.ylabel("Salary (₹)")

plt.show()
````
