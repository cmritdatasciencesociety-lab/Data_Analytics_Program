# What is SQL?

**SQL (Structured Query Language)** is a programming language used to communicate with relational databases.

Using SQL, we can:

- Store data
- Retrieve data
- Update data
- Delete data
- Analyze data

Think of SQL as the language that allows us to ask questions to a database and receive the required information.

---

# Why Do We Need SQL?

Imagine an online shopping company.

Every day, millions of customers:

- Place orders
- Make payments
- Write reviews
- Browse products

All of this information must be stored somewhere.

If the company stores everything in Excel files, it quickly becomes difficult to:

- Manage millions of records
- Search information quickly
- Update records
- Maintain relationships between different datasets

This is why companies use databases.

SQL is the language used to interact with those databases.

---

# SQL in Data Analytics

As Data Analysts, we rarely create applications.

Instead, our job is to answer business questions.

For example:

- How many orders were placed this month?
- Which product generated the highest revenue?
- Which city has the most customers?
- What is the average customer rating?
- Which department has the highest salary expense?

These questions are answered by writing SQL queries.

---

# Real-World SQL Workflow

A typical data analytics workflow looks like this:

```text
Business Problem
        ↓
Data Stored in Database
        ↓
Write SQL Query
        ↓
Retrieve Required Data
        ↓
Analyze Data
        ↓
Generate Business Insights
        ↓
Support Business Decisions
```

SQL is the first step in accessing business data.

---

# Where is SQL Used?

SQL is used in almost every industry that stores data.

Examples include:

| Industry | Example |
|----------|----------|
| E-Commerce | Customers, Orders, Products |
| Banking | Transactions, Accounts |
| Healthcare | Patients, Doctors, Appointments |
| Education | Students, Courses, Marks |
| Social Media | Users, Posts, Likes |
| Logistics | Deliveries, Warehouses, Routes |

Regardless of the industry, if data is stored in a relational database, SQL is used to retrieve and analyze it.

---

# SQL vs Excel vs Pandas

All three can work with data, but they serve different purposes.

| Feature | Excel | SQL | Pandas |
|----------|------|-----|---------|
| Data Storage | Small | Very Large | Reads Existing Data |
| Data Retrieval | Manual | Query Based | Python Code |
| Handles Millions of Rows | No | Yes | Yes (Memory Dependent) |
| Used By | Business Users | Data Professionals | Data Analysts & Data Scientists |

Think of them as complementary tools rather than competitors.

---

---

# What is a Database?

A **database** is an organized collection of related data stored electronically.

It allows businesses to store, manage, retrieve, and update information efficiently.

### Examples

- Customer Database
- Employee Database
- Hospital Database
- Banking Database
- Student Database

---

# What is a Table?

A database contains one or more **tables**.

A table stores information about one specific entity.

Examples:

- Customers
- Orders
- Products
- Employees

Think of a table as similar to an Excel worksheet, where information is organized into rows and columns.

---

# Understanding Rows and Columns

Consider the following Employee table.

| Employee_ID | Name | Department | Salary |
|-------------|------|------------|--------|
| 101 | Alice | HR | 50000 |
| 102 | Bob | Sales | 60000 |
| 103 | Charlie | IT | 75000 |

### Column

A **column** represents an attribute or property.

Examples:

- Employee_ID
- Name
- Department
- Salary

---

### Row

A **row** represents one complete record.

Example:

| Employee_ID | Name | Department | Salary |
|-------------|------|------------|--------|
| 101 | Alice | HR | 50000 |

This entire horizontal entry is one row.

---

# Record vs Field

### Record

A **record** is one complete row.

Example:

| Employee_ID | Name | Department | Salary |
|-------------|------|------------|--------|
| 102 | Bob | Sales | 60000 |

This entire row is one record.

---

### Field

A **field** is one individual value inside a record.

Example:

| Field | Value |
|--------|-------|
| Name | Alice |
| Salary | 50000 |
| Department | HR |

---

# What is a DBMS?

**DBMS** stands for **Database Management System**.

A DBMS is software used to create, manage, store, and retrieve data from databases.

It provides features like:

- Data storage
- Data retrieval
- Data updates
- Data security
- Backup and recovery

Examples:

- Microsoft Access
- MongoDB
- SQLite
- Oracle Database

---

# What is an RDBMS?

**RDBMS** stands for **Relational Database Management System**.

An RDBMS stores data in **tables** and allows relationships between those tables.

Example:

Customers Table

| Customer_ID | Name |
|--------------|------|
| 1 | Alice |
| 2 | Bob |

Orders Table

| Order_ID | Customer_ID | Amount |
|-----------|-------------|--------|
| 101 | 1 | 1500 |
| 102 | 2 | 2500 |

The **Customer_ID** connects the two tables.

This relationship is the foundation of relational databases.

---

# DBMS vs RDBMS

| Feature | DBMS | RDBMS |
|----------|------|--------|
| Stores Data | Yes | Yes |
| Uses Tables | Sometimes | Yes |
| Supports Relationships | No | Yes |
| Follows Relational Model | No | Yes |

For Data Analytics, we primarily work with **RDBMS**.

---

# CRUD Operations

Every database application performs four basic operations.

| Operation | Meaning |
|------------|---------|
| Create | Add new data |
| Read | Retrieve existing data |
| Update | Modify existing data |
| Delete | Remove data |

This is commonly known as **CRUD**.

---

Imagine an Employee database.

### Create

Add a new employee.

↓

### Read

View employee information.

↓

### Update

Change the employee's salary.

↓

### Delete

Remove an employee's record.

These are the four most common database operations.

---

# SQL Command Categories

SQL commands are grouped based on their purpose.

---

## 1. DDL (Data Definition Language)

Used to define or modify database structures.

Examples:

- CREATE
- ALTER
- DROP
- TRUNCATE

---

## 2. DML (Data Manipulation Language)

Used to modify data inside tables.

Examples:

- INSERT
- UPDATE
- DELETE

---

## 3. DQL (Data Query Language)

Used to retrieve data.

Example:

- SELECT

---

## 4. DCL (Data Control Language)

Used to control database permissions.

Examples:

- GRANT
- REVOKE

*For this bootcamp, we will not focus on DCL.*

---

## 5. TCL (Transaction Control Language)

Used to manage transactions.

Examples:

- COMMIT
- ROLLBACK
- SAVEPOINT

*For this bootcamp, we will not focus on TCL.*

---

# SQL Command Categories Summary

| Category | Purpose | Examples |
|-----------|----------|-----------|
| DDL | Define database structure | CREATE, ALTER, DROP |
| DML | Modify data | INSERT, UPDATE, DELETE |
| DQL | Retrieve data | SELECT |
| DCL | Control permissions | GRANT, REVOKE |
| TCL | Manage transactions | COMMIT, ROLLBACK |

# Constraints

Constraints are rules applied to columns.

They help maintain data accuracy.

Examples include:

- PRIMARY KEY
- NOT NULL
- UNIQUE
- DEFAULT

Today we will learn the most important constraint:

**PRIMARY KEY**

---

# PRIMARY KEY

A PRIMARY KEY uniquely identifies every record in a table.

Example:

| Employee_ID | Name |
|--------------|------|
| 101 | Alice |
| 102 | Bob |
| 103 | Charlie |

Every Employee_ID is unique.

Therefore, it can become the PRIMARY KEY.

### Rules

- No duplicate values
- One value for each row
- Used to identify records uniquely

---

# CREATE TABLE Statement

Now let's create our first table.

---

## Syntax

```sql
CREATE TABLE table_name (

    column1 DATA_TYPE,
    column2 DATA_TYPE,
    column3 DATA_TYPE

);
```

---

## Syntax with PRIMARY KEY

```sql
CREATE TABLE table_name (

    id INT PRIMARY KEY,
    column1 DATA_TYPE,
    column2 DATA_TYPE

);
```

# Introduction

In the previous chapter, we created the structure of a table.

However, our table is still empty.

A table becomes useful only after it contains data.

To add records into a table, SQL provides the **`INSERT INTO`** statement.

---

# What is INSERT INTO?

The `INSERT INTO` statement is used to add new records into an existing table.

Every time a new employee joins a company, a new customer registers, or a new order is placed, a new record is inserted into the corresponding table.

---

# Business Example

Imagine a company has created an **Employees** table.

Initially, the table looks like this.

| Employee_ID    | Name | Department | Salary |
| -------------- | ---- | ---------- | ------ |
| *(No Records)* |      |            |        |

The structure exists, but no employee information has been stored yet.

Using `INSERT INTO`, we can start adding employee records.

---

# INSERT INTO Syntax

## Syntax 1 (Recommended)

```sql
INSERT INTO table_name
(column1, column2, column3, ...)
VALUES
(value1, value2, value3, ...);
```

---

## Syntax 2

```sql
INSERT INTO table_name
VALUES
(value1, value2, value3, ...);
```

> This syntax works only when values are provided in the exact order of the table columns.

---

# Why Specify Column Names?

Specifying column names makes the query:

* Easier to read.
* Less error-prone.
* Flexible if the table structure changes.

For this course, we will always follow this approach.

---

# Live Demonstration

🎯 Open the SQL Command Line.

Create an Employees table (if not already created) and insert your first employee record.

---

# Inserting Multiple Records

Instead of inserting one record at a time, SQL allows us to insert multiple records in a single query.

This is useful when importing or creating sample datasets.

---

# Best Practices

* Always specify column names.
* Keep the order of values the same as the columns.
* Insert meaningful sample data.
* Verify data after inserting records.

---

# Common Mistakes

❌ Missing commas between values.

❌ Incorrect order of values.

❌ Missing quotation marks for text values.

❌ Inserting fewer or more values than the number of columns.

❌ Forgetting the semicolon (`;`).

✅ Created a table.

✅ Inserted records into the table.

Now it's time to retrieve the stored data.

In SQL, retrieving data is called **Querying**.

The most commonly used SQL command for querying data is **SELECT**.

---

# What is SELECT?

The `SELECT` statement is used to retrieve data from a table.

It allows us to view:

- All records
- Specific columns
- Required information

Every SQL analysis starts with a `SELECT` query.

---

# Business Example

Imagine you work as a Data Analyst.

Your manager asks,

> "Show me all employee records."

To answer this question, we use the `SELECT` statement.

---

# SELECT Syntax

## Retrieve All Columns

```sql
SELECT *
FROM table_name;
```

`*` means **All Columns**.

---

## Retrieve Specific Columns

```sql
SELECT column1, column2
FROM table_name;
```

This returns only the selected columns.

---

# Live Example

## Display All Employees

```sql
SELECT *
FROM Employees;
```

---

## Display Employee Names

```sql
SELECT Name
FROM Employees;
```

---

## Display Employee Name and Salary

```sql
SELECT Name, Salary
FROM Employees;
```

---

# Understanding the Output

Suppose the Employees table contains:

| Employee_ID | Name | Department | Salary |
|--------------|------|------------|---------|
| 101 | Alice | HR | 45000 |

### Query

```sql
SELECT Name, Salary
FROM Employees;
```

### Output

|------|---------|
| Alice | 45000 |
| Bob | 55000 |
| Charlie | 65000 |

Only the selected columns are displayed.

---

# Introduction

In the previous chapter, we learned how to retrieve data using the `SELECT` statement.

However, in real-world scenarios, we rarely need **all** the records.

For example,

- Show employees whose salary is greater than ₹50,000.
- Display customers from Hyderabad.
- Find products belonging to the Electronics category.

Instead of displaying the entire table, we can filter the required records.

SQL provides the **WHERE** clause for this purpose.

---

# What is WHERE?

The **WHERE** clause is used to filter rows that satisfy a given condition.

It returns only those records that match the specified condition.

---

# Business Example

Suppose the Employees table contains the following data.

| Employee_ID | Name | Department | Salary |
|--------------|------|------------|---------|
| 101 | Alice | HR | 45000 |
| 102 | Bob | Sales | 55000 |
| 103 | Charlie | IT | 65000 |
| 104 | David | HR | 48000 |

If the manager asks:

> "Show employees whose salary is greater than ₹50,000."

We don't need every employee.

We only need employees satisfying that condition.

This is where the **WHERE** clause is used.

---

# WHERE Syntax

```sql
SELECT column_name
FROM table_name
WHERE condition;
```

---

# Retrieve All Columns

```sql
SELECT *
FROM Employees
WHERE Salary > 50000;
```

---

# Retrieve Specific Columns

```sql
SELECT Name, Salary
FROM Employees
WHERE Salary > 50000;
```

---

# Comparison Operators

Comparison operators help us define conditions.

| Operator | Meaning | Example |
|----------|---------|----------|
| = | Equal To | Salary = 50000 |
| > | Greater Than | Salary > 50000 |
| < | Less Than | Salary < 50000 |
| >= | Greater Than or Equal To | Salary >= 50000 |
| <= | Less Than or Equal To | Salary <= 50000 |
| != or <> | Not Equal To | Department != 'HR' |

---

# Chapter 1: Logical Operators (AND, OR, NOT)

---

# Learning Objectives

By the end of this chapter, you will be able to:

- Retrieve records matching either condition using `OR`.

---

# Introduction

In the previous chapter, we filtered records using a **single condition**.

For example,

- Employees earning more than ₹50,000.
- Employees working in the HR department.

But in real-world business problems, one condition is often not enough.

For example,

- Show employees working in the HR department **and** earning more than ₹50,000.
- Show employees from Hyderabad **or** Chennai.
- Show employees **not** working in the Sales department.

To solve these types of problems, SQL provides **Logical Operators**.

---

# What are Logical Operators?

Logical operators are used to combine or modify multiple conditions in a SQL query.

The three most commonly used logical operators are:

- AND
- OR
- NOT

---

# Business Dataset

Assume the Employees table contains the following records.

| Employee_ID | Name | Department | City | Salary |
|--------------|------|------------|------|---------|
| 101 | Alice | HR | Hyderabad | 45000 |
| 102 | Bob | Sales | Chennai | 55000 |
| 103 | Charlie | IT | Bengaluru | 72000 |
| 104 | David | HR | Hyderabad | 48000 |
| 105 | Emma | Finance | Mumbai | 68000 |
| 106 | Frank | Sales | Delhi | 51000 |
| 107 | Grace | IT | Chennai | 76000 |
| 108 | Henry | Finance | Hyderabad | 62000 |

---

# AND Operator

The **AND** operator returns only those rows where **all conditions are true**.

---

## Syntax

```sql
SELECT column_name
FROM table_name
WHERE condition1 AND condition2;
```

---

## Example

Display employees working in the **IT department** and earning more than **₹70,000**.

```sql
SELECT *
FROM Employees
WHERE Department = 'IT'
AND Salary > 70000;
```

---

### Another Example

Display employees from **Hyderabad** whose salary is greater than **₹45,000**.

```sql
SELECT *
FROM Employees
WHERE City = 'Hyderabad'
AND Salary > 45000;
```

---

# OR Operator

The **OR** operator returns rows if **at least one condition is true**.

---

## Syntax

```sql
SELECT column_name
FROM table_name
WHERE condition1 OR condition2;
```

---

## Example

Display employees working in **HR** or **Finance**.

```sql
SELECT *
FROM Employees
WHERE Department = 'HR'
OR Department = 'Finance';
```

---

### Another Example

Display employees from **Mumbai** or **Delhi**.

```sql
SELECT *
FROM Employees
WHERE City = 'Mumbai'
OR City = 'Delhi';
```

---

# NOT Operator

The **NOT** operator returns rows that **do not satisfy** the given condition.

---

## Syntax

```sql
SELECT column_name
FROM table_name
WHERE NOT condition;
```

---

## Example

Display employees who are **not** in the Sales department.

```sql
SELECT *
FROM Employees
WHERE NOT Department = 'Sales';
```

---

### Another Example

Display employees who are **not** from Hyderabad.

```sql
SELECT *
FROM Employees
WHERE NOT City = 'Hyderabad';
```

---

# Difference Between AND and OR

| AND | OR |
|------|----|
| All conditions must be true | At least one condition must be true |
| Returns fewer records | Returns more records |
| Used for stricter filtering | Used for broader filtering |

# Chapter 2: Pattern Matching Using LIKE

---

# Learning Objectives

By the end of this chapter, you will be able to:

- Use the `%` wildcard.
- Use the `_` wildcard.

---

# Introduction

So far, we have filtered data using exact values.

For example,

- Department = 'HR'
- Name = 'Alice'

But what if we don't know the complete value?

Suppose we only know:

- The employee's name starts with **A**
- The employee's name ends with **e**
- The department contains the letter **a**

In such situations, using the **Equal To (`=`)** operator is not sufficient.

SQL provides the **LIKE** operator to search for patterns in text.

---

# What is LIKE?

The **LIKE** operator is used to search for a specific pattern in a text column.

It is commonly used with:

- Employee Names
- Departments
- Product Names
- Customer Names

---

# Wildcards

The LIKE operator uses special characters called **Wildcards**.

| Wildcard | Meaning |
|-----------|---------|
| % | Represents zero or more characters |
| _ | Represents exactly one character |

---

# Using %

The `%` wildcard can match any number of characters.

---

## Starts With

Display employees whose names start with **A**.

```sql
SELECT *
FROM Employees
WHERE Name LIKE 'A%';
```

---

## Ends With

Display employees whose names end with **e**.

```sql
SELECT *
FROM Employees
WHERE Name LIKE '%e';
```

---

## Contains

Display employees whose department contains the letter **a**.

```sql
SELECT *
FROM Employees
WHERE Department LIKE '%a%';
```

---

# Using _

The `_` wildcard represents exactly **one character**.

---

## Example

Display employees whose names have **five letters**.

```sql
SELECT *
FROM Employees
WHERE Name LIKE '_____';
```

Each underscore represents one character.

---

## Another Example

Display employees whose names start with **A** and contain exactly five letters.

```sql
SELECT *
FROM Employees
WHERE Name LIKE 'A____';
```

---

# Understanding the Output

Suppose the following names exist.

| Name |
|------|
| Alice |
| Bob |
| Charlie |
| David |
| Emma |
| Frank |
| Grace |
| Henry |

Query

```sql
SELECT *
FROM Employees
WHERE Name LIKE 'A%';
```

Output

| Name |
|------|
| Alice |

Only names starting with **A** are returned.

# Introduction

In the previous chapter, we learned how to filter data within a range using the `BETWEEN` operator.

Sometimes, instead of a range, we need to search for **multiple specific values**.

For example,

- Employees working in the HR or IT departments.
- Employees with IDs 101, 105, and 108.
- Employees working in the Sales, Finance, or HR departments.

Writing multiple `OR` conditions works, but it makes the query longer.

SQL provides the **IN** operator to simplify such queries.

---

# What is IN?

The **IN** operator is used to check whether a value exists in a given list.

Instead of writing multiple `OR` conditions, we can simply provide a list of values.

---

# Why Use IN?

Without `IN`, we write:

```sql
SELECT *
FROM Employees
WHERE Department = 'HR'
OR Department = 'IT'
OR Department = 'Finance';
```

Using `IN`, the same query becomes:

```sql
SELECT *
FROM Employees
WHERE Department IN ('HR', 'IT', 'Finance');
```

The query is shorter, cleaner, and easier to understand.

---

# IN Syntax

```sql
SELECT column_name
FROM table_name
WHERE column_name IN (value1, value2, value3);
```

# Introduction

So far, we have learned how to:

- Retrieve data using `SELECT`
- Filter data using `WHERE`
- Search patterns using `LIKE`
- Filter ranges using `BETWEEN`
- Filter multiple values using `IN`

However, the retrieved records may not always appear in the order we need.

For example,

- Show employees with the highest salary first.
- Display employee names alphabetically.
- Arrange salaries from lowest to highest.

To organize query results, SQL provides the **ORDER BY** clause.

---

# What is ORDER BY?

The **ORDER BY** clause is used to sort the result of a query.

You can sort data:

- In Ascending Order (`ASC`)
- In Descending Order (`DESC`)

---

# ORDER BY Syntax

```sql
SELECT column_name
FROM table_name
ORDER BY column_name;
```

By default, SQL sorts data in **Ascending (ASC)** order.

---

# Ascending Order (ASC)

Ascending order arranges data from:

- Smallest to Largest (Numbers)
- A to Z (Text)

---

## Example

Display employees sorted by salary in ascending order.

```sql
SELECT *
FROM Employees
ORDER BY Salary ASC;
```

---

## Example

Display employees sorted alphabetically by name.

```sql
SELECT *
FROM Employees
ORDER BY Name ASC;
```

---

# Descending Order (DESC)

Descending order arranges data from:

- Largest to Smallest (Numbers)
- Z to A (Text)

---

## Example

Display employees with the highest salary first.

```sql
SELECT *
FROM Employees
ORDER BY Salary DESC;
```

---

## Example

Display employee names from Z to A.

```sql
SELECT *
FROM Employees
ORDER BY Name DESC;
```

---

# Sorting by Multiple Columns

SQL also allows sorting using more than one column.

The first column is sorted first.

If multiple rows have the same value, SQL sorts them using the second column.

---

## Example

Sort employees by Department and then by Salary.

```sql
SELECT *
FROM Employees
ORDER BY Department ASC, Salary DESC;
```

# Why Do We Need UPDATE?

Business data changes over time.

For example,

- An employee receives a salary increment.
- An employee changes departments.
- An employee's name is corrected.

Instead of deleting and inserting the record again, we simply update the existing record.

SQL provides the **UPDATE** statement for this purpose.

---

# UPDATE Statement

General Syntax

```sql
UPDATE table_name
SET column_name = value
WHERE condition;
```

> **Important:** Always use the `WHERE` clause unless you intentionally want to update every record in the table.

---

# Example 1: Update Employee Salary

Suppose employee **Bob (Employee_ID = 102)** receives a salary increment.

```sql
UPDATE Employees
SET Salary = 60000
WHERE Employee_ID = 102;
```

---

# Example 2: Update Employee Department

David has been transferred from **HR** to **Finance**.

```sql
UPDATE Employees
SET Department = 'Finance'
WHERE Employee_ID = 104;
```

---

# Example 3: Update Multiple Columns

Frank has been transferred to the **IT** department and his salary has been increased.

```sql
UPDATE Employees
SET Department = 'IT',
    Salary = 58000
WHERE Employee_ID = 106;
```

---

# Example 4: Update Multiple Records

The company has renamed the **HR** department to **Human Resources**.

```sql
UPDATE Employees
SET Department = 'Human Resources'
WHERE Department = 'HR';
```

---

# Why Do We Need DROP TABLE?

Sometimes a table is no longer required.

For example:

- The table was created incorrectly.
- A new table structure needs to be designed.
- The table is no longer used in the project.

Instead of keeping unnecessary tables, we can permanently remove them using the **DROP TABLE** statement.

---

# DROP TABLE Statement

General Syntax

```sql
DROP TABLE table_name;
```

---

# Example

Remove the **Employees** table from the database.

```sql
DROP TABLE Employees;
```

After executing this statement,

- The table is permanently removed.
- All records stored inside the table are also deleted.
- The table cannot be queried unless it is created again.

---

# Verify the Result

Display all available tables.

```sql
.tables
```

or

```sql
SELECT name
FROM sqlite_master
WHERE type='table';
```

If the **Employees** table is not displayed, it has been successfully removed.

---

# Important Note

⚠️ `DROP TABLE` permanently removes both:

- The table structure.
- All data stored inside the table.

This operation **cannot be undone**.

Always verify that you no longer need the table before executing the command.

---

# DROP vs DELETE

| DROP TABLE | DELETE |
|------------|--------|
| Removes the entire table | Removes records from a table |
| Table structure is deleted | Table structure remains |
| Data is deleted | Data is deleted |
| Table must be created again to use it | The table can still be used |
