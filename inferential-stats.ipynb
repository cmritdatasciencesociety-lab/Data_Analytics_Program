# What is Inferential Statistics?

Inferential Statistics is the branch of statistics that helps us make conclusions or predictions about an entire population using only a sample of data.

Instead of collecting data from every individual, we analyze a smaller representative sample and use statistical techniques to make informed decisions.

In simple words,

> **Inferential Statistics = Making decisions about a large population using a small sample.**

---

# Why Do We Need Inferential Statistics?

In many real-world situations, collecting data from every individual is:

- Expensive
- Time-consuming
- Sometimes impossible

For example,

Suppose Amazon wants to know the average delivery satisfaction of all its customers.

Would it be practical to ask every customer?

**No.**

Instead, Amazon collects feedback from a sample of customers and uses Inferential Statistics to estimate the satisfaction level of all customers.

This saves:

- Time
- Money
- Resources

while still providing reliable insights.

---

# Descriptive Statistics vs Inferential Statistics

| Descriptive Statistics | Inferential Statistics |
|-------------------------|-------------------------|
| Summarizes existing data | Makes conclusions from data |
| Describes what happened | Predicts or estimates what is likely true |
| Works with the available dataset | Uses a sample to understand a population |
| Uses charts, averages, and summaries | Uses probability and statistical tests |
| Example: Average salary of 500 employees | Example: Estimate average salary of all employees in the company |

---

# Business Example

Imagine an online shopping company launches a new website design.

They ask **500 customers** to rate the new experience.

The average rating from those 500 customers is **4.6 out of 5**.

Using Inferential Statistics, the company estimates that **most customers** are likely satisfied with the new website, even though millions of customers use it.

This decision is made using only a sample instead of surveying everyone.

# Why Do We Need Sampling?

Imagine you are working as a Data Analyst at an e-commerce company.

The company wants to know the average customer satisfaction of **10 million customers**.

Would it be practical to ask every customer?

**No!**

Collecting data from every individual would require:

- More time
- More money
- More resources

Instead, we collect data from a smaller group that represents the entire population.

This process is called **Sampling**.

---

# What is Population?

A **Population** is the complete collection of individuals, objects, or observations that we want to study.

In simple words,

> **Population = The entire group we are interested in.**

### Examples

| Study | Population |
|--------|------------|
| Student Performance | All students in the college |
| Customer Satisfaction | All customers of Amazon |
| Election Survey | All eligible voters |
| Product Quality | Every product manufactured today |

---

# What is a Sample?

A **Sample** is a smaller group selected from the population.

The sample should represent the population as closely as possible.

In simple words,

> **Sample = A small portion of the population.**

### Examples

| Population | Sample |
|------------|--------|
| 10,000 students | 500 students |
| 5 million customers | 2,000 customers |
| All voters in India | 5,000 surveyed voters |
| 50,000 manufactured products | 300 inspected products |

---

# Population vs Sample

| Population | Sample |
|------------|--------|
| Entire group | Small subset |
| Usually very large | Smaller and manageable |
| Expensive to collect | Less expensive |
| Time-consuming | Faster to collect |
| Gives exact values | Used to estimate population values |

---

# Parameter vs Statistic

These two terms are often confused.

## Parameter

A **Parameter** is a numerical value calculated from the **Population**.

Examples:

- Population Mean
- Population Standard Deviation
- Population Proportion

Since collecting data from the entire population is difficult, parameters are often unknown.

---

## Statistic

A **Statistic** is a numerical value calculated from a **Sample**.

Examples:

- Sample Mean
- Sample Standard Deviation
- Sample Proportion

Statistics are used to estimate population parameters.

---

# Parameter vs Statistic

| Parameter | Statistic |
|-----------|-----------|
| Calculated from Population | Calculated from Sample |
| Usually unknown | Can be calculated |
| Represents the whole population | Represents only the sample |

---

# What is Sampling?

Sampling is the process of selecting a subset of individuals from a population for analysis.

The goal of sampling is to make reliable conclusions about the entire population while saving:

- Time
- Cost
- Effort

---

# Sampling Techniques

There are many sampling methods.

In this course, we will learn four commonly used techniques.

---

# 1. Simple Random Sampling

Every individual has an equal chance of being selected.

### Example

A college has 5,000 students.

A computer randomly selects 300 students.

Every student has an equal probability of being chosen.

### Advantages

- Easy to understand
- Reduces selection bias
- Fair selection

---

# 2. Stratified Sampling

The population is divided into groups (called strata).

Samples are then selected from each group.

### Example

Suppose a college has:

- 60% Boys
- 40% Girls

To maintain the same proportion in the sample, we randomly select students from both groups.

This makes the sample more representative.

### Advantages

- Better representation
- Useful when groups differ significantly

---

# 3. Systematic Sampling

Individuals are selected at regular intervals.

### Example

Suppose there are 10,000 customers.

Instead of choosing randomly, select every 50th customer.

Examples:

50th

100th

150th

200th

...

### Advantages

- Simple
- Fast
- Easy to implement

---

# 4. Convenience Sampling

Samples are selected based on availability or ease of access.

### Example

A teacher asks only students sitting in the first two rows to answer a survey.

### Advantages

- Very fast
- Easy to collect

### Disadvantage

May not represent the entire population.

---

# Comparison of Sampling Techniques

| Technique | Selection Method | Best For |
|-----------|------------------|-----------|
| Simple Random | Completely Random | General surveys |
| Stratified | Random within groups | Population with different categories |
| Systematic | Fixed intervals | Large ordered datasets |
| Convenience | Easily available individuals | Quick surveys |

---

# Key Takeaways

- Population is the complete group we want to study.
- A Sample is a subset of the population.
- Parameters describe populations.
- Statistics describe samples.
- Sampling helps save time and cost.
- Different sampling techniques are used depending on the business problem.

---

# What is a Distribution?

A **Distribution** describes how data values are spread.

It tells us:

- Where most values are located.
- Whether values are concentrated or spread out.
- Whether there are extreme values (outliers).

In simple words,

> **A distribution shows how data is distributed across different values.**

---

# Example

Suppose we record the marks of 100 students.

Some students score:

- Very high marks
- Most students score around the average
- Some students score very low

A distribution helps us visualize this pattern.

---

# What is a Normal Distribution?

A **Normal Distribution**, also called the **Gaussian Distribution**, is one of the most common probability distributions.

It forms a **bell-shaped curve**, where:

- Most observations lie near the center.
- Fewer observations lie far from the center.
- The left and right sides are symmetrical.

It is represented as a smooth, bell-shaped curve.

---

# Characteristics of a Normal Distribution

A Normal Distribution has the following properties:

- Bell-shaped curve
- Symmetrical about the center
- Mean = Median = Mode
- Most observations lie near the center
- Fewer observations occur as we move away from the center
- The total area under the curve equals 1 (or 100%)

---

# Bell Curve

A Normal Distribution is often called a **Bell Curve** because of its shape.

```text
                *
             *     *
           *         *
         *             *
       *                 *
     *                     *
   *                         *
 *                             *
-------------------------------------------->
```

The highest point of the curve represents the value where most observations occur.

---

# Mean, Median and Mode

In a perfectly Normal Distribution,

```text
Mean = Median = Mode
```

All three measures are located exactly at the center of the distribution.

---

# The Empirical Rule (68–95–99.7 Rule)

The Empirical Rule describes how data is distributed in a Normal Distribution.

### About 68% of data

lies within **1 Standard Deviation** from the Mean.

---

### About 95% of data

lies within **2 Standard Deviations** from the Mean.

---

### About 99.7% of data

lies within **3 Standard Deviations** from the Mean.

---

## Summary

| Range | Percentage of Data |
|--------|--------------------:|
| Mean ± 1 Standard Deviation | 68% |
| Mean ± 2 Standard Deviations | 95% |
| Mean ± 3 Standard Deviations | 99.7% |

---

# Real-Life Examples of Normal Distribution

Many real-world datasets approximately follow a Normal Distribution.

Examples include:

- Students' heights
- IQ scores
- Blood pressure
- Birth weights
- Measurement errors
- Exam scores (in many large exams)

Not every dataset follows a Normal Distribution, but many statistical methods assume that the data is approximately normal.

---

# Why is Normal Distribution Important?

Many statistical techniques are built on the assumption that data follows a Normal Distribution.

Examples include:

- Confidence Intervals
- Hypothesis Testing
- Z-Test
- T-Test

Understanding the Normal Distribution helps us make reliable statistical decisions.

---

# Business Example

Suppose a company tracks the daily delivery time of thousands of orders.

Most deliveries take around **3 days**.

Some deliveries are slightly faster.

Some deliveries are slightly slower.

Very few deliveries take an unusually long time.

If the delivery times follow a Normal Distribution, analysts can estimate:

- Typical delivery time
- Unusual delays
- Expected customer experience

---

# Key Takeaways

- A Distribution shows how data is spread.
- A Normal Distribution is a bell-shaped, symmetrical distribution.
- Mean, Median and Mode are equal in a Normal Distribution.
- Most observations lie near the center.
- The Empirical Rule explains how data is distributed.
- Many inferential statistical methods assume normally distributed data.

# Why Can't We Trust Every Sample?

Every sample is different.

For example,

Sample 2:

Sample 3:

Average Rating = **4.4**

Sample 4:

Average Rating = **4.3**

Different samples produce different results.

Inferential Statistics helps us decide whether these differences are meaningful or simply due to random variation.
